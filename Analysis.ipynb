Define some helper functions we'll use to clean up the data by removing values from fields that are of the improper type.

```python
def isdate(value):
    try:
        pd.to_datetime(value).date()
        return True
    except:
        return False
    
def isfloat(value):
    try:
        np.float64(value)
        return True
    except:
        return False
    
def isstring(value):
    try:
        str(value)
        return True
    except:
        return False
    
def isvalidname(value):
    try:
        return ' ' not in value
    except:
        return False
```

Define a function for extracting the backbone of a graph that was developed by [Serrano, Bovuna, and Vespigani](http://www.pnas.org/content/106/16/6483.abstract) and re-implemented by [me](https://gist.github.com/brianckeegan/8846206).

```python
def extract_backbone(g, alpha):
    backbone_graph = nx.Graph()
    for node in g:
        k_n = len(g[node])
        if k_n > 1:
            sum_w = sum( g[node][neighbor]['weight'] for neighbor in g[node] )
            for neighbor in g[node]:
                edgeWeight = g[node][neighbor]['weight']
                pij = float(edgeWeight)/sum_w
                if (1-pij)**(k_n-1) < alpha: # equation 2
                    backbone_graph.add_edge( node,neighbor, weight = edgeWeight)
    return backbone_graph
```

# Data cleanup and groupbys

Subset the data to only include tweets that are retweets, then cleanup the data. Groupby `user_screen_name` and `retweeted_user_screen_name` to create an edgelist, then perform an aggregate function on the groupby object to extract additional metadata. Perform more cleanup and do some additional calculations to create edge-level attributes.

```python
# Subset data to only include retweets
rt_data = data.dropna(subset=['retweeted_user_screen_name'])

# Subset data to remove non-float-able values, convert remainder to floats
rt_data = rt_data[rt_data['retweeted_count'].apply(isfloat)]
rt_data['retweeted_count'] = rt_data['retweeted_count'].apply(np.float64)

# Groupby user_name and retweeted_username that will become edgelist
rt_gb = rt_data.groupby(['user_screen_name','retweeted_user_screen_name'])

# Create rt_agg DataFrame with aggregate statistics for each edge
rt_agg = rt_gb.agg({'tweet_timestamp':{'ts_min':min,'ts_max':max},
                    'tweet_id':{'weight':len},
                    'retweeted_count':{'rt_max':max,'rt_min':min,'rt_sum':np.sum}
                    })
rt_agg.columns = rt_agg.columns.droplevel(0)

# Include only edges with >1 RT
#rt_agg = rt_agg[rt_agg['weight'] > 1]

# Data cleanup
rt_agg['ts_min'] = (rt_agg['ts_min'] - pd.Timestamp('2014-10-21'))/np.timedelta64(1,'D')
rt_agg['ts_max'] = (rt_agg['ts_max'] - pd.Timestamp('2014-10-21'))/np.timedelta64(1,'D')
rt_agg['rt_fraction'] = rt_agg['weight']/rt_agg['rt_sum']
rt_agg.head()
```

Subset the data to include all tweets made by users who did a retweet in the previous step. Do additional data cleanup, aggregation, and computation to create node-level attributes.

```python
# Subset data to only include tweets for users in graph
tweeters = data[data['user_screen_name'].isin(rt_agg.reset_index()['user_screen_name'].values)]
retweeters = data[data['user_screen_name'].isin(rt_agg.reset_index()['retweeted_user_screen_name'].values)]
g_data = pd.concat([tweeters,retweeters])

# Remove rows where account creation date is not valid
g_data = g_data[g_data['user_created_at'].apply(isdate)]
g_data['user_created_at'] = g_data['user_created_at'].apply(pd.Timestamp)

# Remove rows where other counts are not valid
g_data = g_data[g_data['user_followers_count'].apply(isfloat)]
g_data['user_followers_count'] = g_data['user_followers_count'].apply(np.float64)

g_data = g_data[g_data['user_friends_count'].apply(isfloat)]
g_data['user_friends_count'] = g_data['user_friends_count'].apply(np.float64)

g_data = g_data[g_data['user_statuses_count'].apply(isfloat)]
g_data['user_statuses_count'] = g_data['user_statuses_count'].apply(np.float64)

# Group by user_screen_name to create nodelist
g_gb = g_data.groupby('user_screen_name')

# Create agggregated DataFrame of all users tweets and summary statistics
g_agg = g_gb.agg({'tweet_id':{'tweets':len},
                  'user_followers_count':{'followers_min':min,'followers_max':max},
                  'user_friends_count':{'followees_min':min,'followees_max':max},
                  'user_statuses_count':{'statuses_min':min,'statuses_max':max},
                  'user_created_at':{'account_age':min}})
g_agg.columns = g_agg.columns.droplevel(0)

g_agg['account_age'] = (pd.Timestamp('2014-10-24') - g_agg['account_age'])/np.timedelta64(1,'D')
g_agg['tweets_delta'] = g_agg['statuses_max'] - g_agg['statuses_min']
g_agg['followers_delta'] = (g_agg['followers_max'] - g_agg['followers_min'])/g_agg['followers_min']
g_agg['followees_delta'] = (g_agg['followees_max'] - g_agg['followees_min'])/g_agg['followees_min']
g_agg['tweet_intensity'] = g_agg['tweets']/(g_agg['tweets_delta']+1)
g_agg['tweet_fraction'] = g_agg['tweets']/g_agg['statuses_min']

g_agg.sort('tweets',ascending=False).head()
```

# Mentions

Define some new helper functions that will extract regular expressions from text and compute the reciprocity of a directed graph. Use a regex to extract a list of mentions and hashtags and store these in the `data` DataFrame.

```python
def regex_matcher(pattern,string):
    try:
        results = re.findall(pattern,string)
        if len(results) > 0:
            return [i.lower() for i in results]
        else:
            return np.nan
    except TypeError:
        return np.nan
    
def reciprocity(graph):
    d = dict()
    for n in graph.nodes_iter():
        s = set(graph.successors(n))
        p = set(graph.predecessors(n))
        if len(s | p) > 0:
            r = len(s & p)/float(len(s | p))
        else:
            r = 0
        d[n] = r
    return d
    
# From http://stackoverflow.com/a/6351873/1574687
_m = r'(?<=^|(?<=[^a-zA-Z0-9-_\.]))@([A-Za-z_]+[A-Za-z0-9_]+)'
_h = r'(?<=^|(?<=[^a-zA-Z0-9-_\.]))#([A-Za-z_]+[A-Za-z0-9_]+)'

# Create list of mentions for each tweet
data['mentions'] = data['tweet_text'].map(lambda x:regex_matcher(_m,x))
data['hashtags'] = data['tweet_text'].map(lambda x:regex_matcher(_h,x))
```

Subset the data to only include tweets that have a valid mention as extracted above. Print out the number of tweets containing mentions in the dataset before the subsequent cleanup steps.

```python
mentions_data = data.dropna(subset=['mentions'])
print len(mentions_data)
mentions_data = mentions_data[mentions_data['user_screen_name'].apply(isvalidname)]
print len(mentions_data)
```

Create an empty DiGraph object `mentions_g` to hold all the mentions relationships, then loop over the subsetted data above to add an edge for each mention in each tweet. User names are not case sensitive, but we want to make sure there are no redundant nodes, so lower-case everything. Update the weights if the edge already appears in the graph, otherwise add a new edge to the graph with weight 1.

Compute some node-level social network features from the graph such as out-degree centrality, in-degree centrality, clustering coefficient, reciprocity, and normalized degree difference.

```python
mg_size = len(mentions_g)

mg_df = pd.DataFrame()
mg_df['odc'] = pd.Series({k:int(v*(mg_size-1)) for k,v in nx.out_degree_centrality(mentions_g).iteritems()})
mg_df['idc'] = pd.Series({k:int(v*(mg_size-1)) for k,v in nx.in_degree_centrality(mentions_g).iteritems()})
mg_df['cc'] = pd.Series(nx.clustering(mentions_g.to_undirected()))
mg_df['reciprocity'] = pd.Series(reciprocity(mentions_g))
mg_df['diff'] = (mg_df['idc'] - mg_df['odc'])/(mg_df['idc'] + mg_df['odc'])
```

# Multigraph analysis

Identify the intersection of edges that appear in both the `mentions_g` and `rt_g` networks. Use this overlap to create a new graph, `multig`. Add in some edge properties based on the weight of retweets or mentions in the previous relationships.

```python
overlap = list(set(mentions_g.edges_iter()) & set(rt_g.edges_iter()))

multig = nx.DiGraph()
multig.add_edges_from(overlap)

for (i,j) in multig.edges_iter():
    m_w = float(mentions_g[i][j]['weight'])
    rt_w = float(rt_g[i][j]['weight'])
    multig.add_edge(i,j,mentions=m_w,retweets=rt_w,total_weight=m_w + rt_w)
```

```python
multig_data = data[data['user_screen_name'].isin(multig.nodes())]

# Remove rows where account creation date is not valid
multig_data = multig_data[multig_data['user_created_at'].apply(isdate)]
multig_data['user_created_at'] = multig_data['user_created_at'].apply(pd.Timestamp)

# Remove rows where other counts are not valid
multig_data = multig_data[multig_data['user_followers_count'].apply(isfloat)]
multig_data['user_followers_count'] = multig_data['user_followers_count'].apply(np.float64)

multig_data = multig_data[multig_data['user_friends_count'].apply(isfloat)]
multig_data['user_friends_count'] = multig_data['user_friends_count'].apply(np.float64)

multig_data = multig_data[multig_data['user_statuses_count'].apply(isfloat)]
multig_data['user_statuses_count'] = multig_data['user_statuses_count'].apply(np.float64)

# Group by user_screen_name to create nodelist
multig_gb = multig_data.groupby('user_screen_name')

# Create agggregated DataFrame of all users tweets and summary statistics
multig_agg = multig_gb.agg({'tweet_id':{'tweets':len},
                  'user_followers_count':{'followers_min':min,'followers_max':max},
                  'user_friends_count':{'followees_min':min,'followees_max':max},
                  'user_statuses_count':{'statuses_min':min,'statuses_max':max},
                  'user_created_at':{'account_age':min}})
multig_agg.columns = multig_agg.columns.droplevel(0)

multig_agg['account_age'] = (pd.Timestamp('2014-10-24') - multig_agg['account_age'])/np.timedelta64(1,'D')
multig_agg['tweets_delta'] = multig_agg['statuses_max'] - multig_agg['statuses_min']
multig_agg['followers_delta'] = multig_agg['followers_max'] - multig_agg['followers_min']
multig_agg['followees_delta'] = multig_agg['followees_max'] - multig_agg['followees_min']
multig_agg['tweet_intensity'] = (multig_agg['tweets']+1)/(multig_agg['tweets_delta']+1)
multig_agg['tweet_fraction'] = multig_agg['tweets']/multig_agg['statuses_min']

# Lowercase usernames for future joinabilitiy
multig_agg.index = multig_agg.reset_index()['user_screen_name'].apply(lambda x:x.lower())
```

Compute SNA statistics for the nodes in the network. Join the SNA and tweet statistics together into the `multi` DataFrame.

```python
multig_size = len(multig)

multig_df = pd.DataFrame()
multig_df['odc'] = pd.Series({k:int(v*(multig_size-1)) for k,v in nx.out_degree_centrality(multig).iteritems()})
multig_df['idc'] = pd.Series({k:int(v*(multig_size-1)) for k,v in nx.in_degree_centrality(multig).iteritems()})
multig_df['cc'] = pd.Series(nx.clustering(multig.to_undirected()))
multig_df['reciprocity'] = pd.Series(reciprocity(multig))
multig_df['diff'] = (multig_df['idc'] - multig_df['odc'])/(multig_df['idc'] + multig_df['odc'])

multi = multig_agg.join(multig_df)
```

Add these node attributes to the graph and save the graph to GEXF for visualization in Gephi.

```python
for n in multig.nodes_iter():
    try:
        # There is some bizzare bug in the aggregate function that is returning DataFrames instead of Series
        if type(multig_agg.ix[n]) == pd.core.frame.Series:
            try:
                d = {k:float(v) for k,v in dict(multig_agg.ix[n]).items()}
                multig.add_node(n,d)
            except TypeError:
                print n
    except KeyError:
        pass

nx.write_gexf(multig,'mentions_retweet_multigraph.gexf')
```
